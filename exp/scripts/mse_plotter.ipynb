```python
#### Notebook code for plotting the MSE results from the power flow experiments. 

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Your data
data = {
    'Quantity': ['V', 'V', 'V', 'V', 'θ', 'θ', 'θ', 'θ', 'Pg', 'Pg', 'Pg', 'Pg', 'Qg', 'Qg', 'Qg', 'Qg'],
    'Grid':     ['IEEE24', 'IEEE39', 'UK', 'IEEE118'] * 4,
    'MAE':      [
        0.1682, 0.0426, 0.0476, 0.2156,   # V
        0.2206, 0.2159, 0.2385, 0.3012,  # θ
        0.1276, 0.0756, 0.1379, 0.2265,   # Pg
        0.2067, 0.1212, 0.3213, 0.3888, # Qg
    ]
}

df = pd.DataFrame(data)

# Plot
fig, ax = plt.subplots(figsize=(9, 5))

sns.barplot(
    data=df,
    x='Quantity',
    y='MAE',
    hue='Grid',
    palette=['#4CAF82', '#E07B3A', '#7B68C8', '#E0417B'],
    ax=ax
)

ax.set_title('Power flow')
ax.set_xlabel('Quantity')
ax.set_ylabel('Mean Absolute Error (MAE)')
ax.legend(title='Grid')

ax.ticklabel_format(axis='y', style='sci', scilimits=(0, 0))

plt.tight_layout()
plt.show()
```
